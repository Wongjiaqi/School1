```python
#第二题
import talib as ta
df_600036=pd.read_excel('sz50.xlsx',sheetname='600036.XSHG',index_col='datetime')
arr=np.array(df_600036.close)
ma10=ta.MA(arr,10)
print(type(ma10))
print(ma10[-5:])
```
