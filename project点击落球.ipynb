```python
#OOPy Animation Examples

#i   Falling circles
import random

#define customed classes

class Circle(object):

    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.r = random.randint(5, 30)
        self.c = random.choice(["salmon","dark slate blue","slate gray",
                                "gold","spring green"])
        self.v = 10
        self.m = random.randint(1, 10)
        self.t = 0

    def __str__(self):
        return str(self.x)+" "+str(self.y)

    def __repr__(self):
        pass

    def __hash__(self):
        pass

    def __eq__(self,other):
        return (self.r == other.r)

    def __gt__(self,other):
        return (self.r > other.r)

    def __ge__(self):
        pass

    def __lt__(self):
        pass

    def __le__(self):
        pass

    def __ne__(self):
        pass

    def __add__(self,other):
        return (self.r + other.r)

    def setPosition(self): #ignore air resistance
        g = 9.6
        self.v += (g*self.t**2)//2
        self.y += self.v

    def recordTime(self):
        self.t += 1

    def draw(self,canvas,data):
        if (self.y-self.r >= data.height): pass
        (x,y,r) = (self.x, self.y, self.r)
        canvas.create_oval(x-r, y-r, x+r, y+r, fill=self.c)

#Actual Animation
from tkinter import *

def init(data):
    data.c = []
    data.timerDelay = 250

def mousePressed(event, data):
    data.c.append(Circle(event.x, event.y))

def keyPressed(event, data):
    pass

def timerFired(data):
    for circle in data.c:
        print(circle)
        if (circle == circle): print("equal")
        circle.recordTime()
        circle.setPosition()

def redrawAll(canvas, data):
    canvas.create_text(data.width/2, data.height*4/5,
                       text="Click to create new falling circles!")
    for circle in data.c:
        circle.draw(canvas, data)

def run(width=300, height=300):
    def redrawAllWrapper(canvas, data):
        canvas.delete(ALL)
        redrawAll(canvas, data)
        canvas.update()    

    def mousePressedWrapper(event, canvas, data):
        mousePressed(event, data)
        redrawAllWrapper(canvas, data)

    def keyPressedWrapper(event, canvas, data):
        keyPressed(event, data)
        redrawAllWrapper(canvas, data)

    def timerFiredWrapper(canvas, data):
        timerFired(data)
        redrawAllWrapper(canvas, data)
        # pause, then call timerFired again
        canvas.after(data.timerDelay, timerFiredWrapper, canvas, data)
    # Set up data and call init
    class Struct(object): pass
    data = Struct()
    data.width = width
    data.height = height
    data.timerDelay = 100 # milliseconds
    init(data)
    # create the root and the canvas
    root = Tk()
    canvas = Canvas(root, width=data.width, height=data.height)
    canvas.pack()
    # set up events
    root.bind("<Button-1>", lambda event:
                            mousePressedWrapper(event, canvas, data))
    root.bind("<Key>", lambda event:
                            keyPressedWrapper(event, canvas, data))
    timerFiredWrapper(canvas, data)
    # and launch the app
    root.mainloop()  # blocks until window is closed
    print("bye!")

run(400,800)
```
